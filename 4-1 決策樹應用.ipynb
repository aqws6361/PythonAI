```python
#匯入套件
from sklearn import tree, datasets
import matplotlib.pyplot as plt
```

載入資料庫

```python
x, y = datasets.load_iris(return_X_y=True) #載入資料
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3) #建立決策樹模型
clf = clf.fit(x, y) #模型訓練
```

```python
plt.figure(figsize=(10, 10)) #圖框放大
tree.plot_tree(clf) #繪製決策樹
plt.show() #圖像輸出
```
